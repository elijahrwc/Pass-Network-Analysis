```python
# Set match_id
MATCH_ID = 3869685

file_path = 'open-data/data/three-sixty/3788741.json'

# Get events by creating a DataFrame
match_events = sb.events(match_id=MATCH_ID)

# Access the 360 data
match_360_df = pd.read_json(file_path) 
```

```python
# Merge the two DataFrames by event_uuid so that we can access the 360 data
df = pd.merge(match_events, match_360_df, left_on='id', right_on='event_uuid', how='left')

#Print all player_ids and their names
df['player_id'].unique()
```

```python
df.columns
```

```python
# Filter for Messi's passes
MESSI_ID = 5503

df = df[(df['player_id'] == MESSI_ID) & (df['type'] == 'Pass')].reset_index(drop=True)
```

```python
df.location
```

```python
# Create a new column for the start location
df[['x_start', 'y_start']] = pd.DataFrame(df.location.tolist(), index=df.index)

# Create a new column for the end location
df[['x_end', 'y_end']] = pd.DataFrame(df.pass_end_location.tolist(), index=df.index)
```

```python
# Set the pitch dimensions
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white')

# Create a figure
fig, ax = pitch.draw()

# Plot the start and end locations
pitch.scatter(df.x_start, df.y_start, s=100, c='blue', ax=ax)
pitch.scatter(df.x_end, df.y_end, s=100, c='red', ax=ax)

# Plot the pass lines
for i in range(len(df)):
    pitch.arrows(df.x_start[i], df.y_start[i], df.x_end[i], df.y_end[i], width=2, color='blue', ax=ax)

# Show the plot
fig.set_size_inches(12, 8)
fig.tight_layout()
fig.show()
```

Create a map of every pass made during that game.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a dictionary to map player IDs to player names
player_names = {
    5503: 'Lionel Messi',
    2995: 'Di María',
    7797: 'R. De Paul',
    20572: 'C. Romero',
    27886: 'Mac Allister',
    29560: 'J. Álvarez',
    29201: 'Molina',
    3090: 'Otamendi',
    38718: 'E. Fernandez',
    19597: 'M. Acuña',
    11456: 'L. Martínez',
    28263: 'Montiel'
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes for each player
for player_id in df['player_id'].dropna().unique():
    G.add_node(player_id)

# Add edges for each pass with weights
for i in range(len(df)):
    if pd.notna(df['player_id'][i]) and pd.notna(df['pass_recipient_id'][i]):
        if G.has_edge(df['player_id'][i], df['pass_recipient_id'][i]):
            G[df['player_id'][i]][df['pass_recipient_id'][i]]['weight'] += 1
        else:
            G.add_edge(df['player_id'][i], df['pass_recipient_id'][i], weight=1)

# Create a layout for the nodes
pos = nx.spring_layout(G)

# Draw the nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=700)
edges = nx.draw_networkx_edges(G, pos, width=[G[u][v]['weight'] for u, v in G.edges()])

# Use the player_names dictionary to label the nodes
nx.draw_networkx_labels(G, pos, labels=player_names, font_size=12, font_color='Orange')

# Show the plot
plt.axis('off')
plt.show()

```

```python
# Get the lineup data and extract jersey numbers
lineup_df = parser.lineup(MATCH_ID)
jersey_data=lineup_df[['player_id','jersey_number']]

# Merge the jersey numbers with the evenmt data
df= pd.merge(event_df,jersey_data,on='player_id')

df
```

```python
# make player_id the jersey number
df['passer'] = df['jersey_number']
passes = df[df['type_name']=='Pass']
successful = passes[passes['outcome_name'].isnull()]
```

```python
# Finalized DataFrame
rec= pd.to_numeric(successful['pass_recipient_id'],downcast='integer')
jersey_data.rename(columns={'player_id':'pass_recipient_id'},inplace=True)
jersey_data.rename(columns={'jersey_number':'pass_recipient'},inplace=True)
successful=pd.merge(df,jersey_data,on='pass_recipient_id')
successful
```

```python
# Get the first substitution minute
subs = event_df[event_df['type_name']=='Substitution']
subs = subs['minute']
firstSub=subs.min()

# Filter for passes before first substitution
successful = successful[successful['minute']<firstSub]
```

```python
# Print variable types
successful.dtypes
```

```python
# Get the average location of each player
average_locations= successful.groupby('passer').agg({'x':['mean'],'y':['mean','count']})
average_locations.columns = ['x','y','count']
average_locations
```

```python
# Make pass_between dataframe
pass_between = successful.groupby(['passer','pass_recipient']).id.count().reset_index()
pass_between.rename({'id':'pass_count'},axis='columns',inplace=True)
pass_between
```

```python
# Merge dataframes pass_between and average_locations
pass_between= pass_between.merge(average_locations,left_on='passer',right_index=True)
pass_between
```

```python
# Merge dataframes pass_between and average_locations on pass_recipient
pass_between= pass_between.merge(average_locations,left_on='pass_recipient',right_index=True,suffixes=['','_end'])
pass_between
```

```python
# Filter out passes with no count, low count, or duplicates
pass_between =pass_between[pass_between['pass_count']>3]
pass_between.drop_duplicates()
pass_between
```
