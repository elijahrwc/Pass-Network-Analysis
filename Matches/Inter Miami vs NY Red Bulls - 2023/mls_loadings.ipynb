```python
# Filter for Messi's passes
MESSI_ID = 5503
messi_passes = event_df[(event_df['player_id'] == MESSI_ID) & (event_df['type_name'] == 'Pass')].reset_index(drop=True)
messi_passes.columns
```

```python
# Filter and reshape messi_passes df
filtered = messi_passes[['player_name', 'player_id', 'x','y','end_x', 'end_y','pass_recipient_name', 'pass_recipient_id', 'outcome_id','outcome_name']]
filtered[filtered['outcome_name'].isnull()]

# Create tuple columns for start and end coordinates
filtered['location'] = list(zip(filtered.x, filtered.y))
filtered['pass_end_location'] = list(zip(filtered.end_x, filtered.end_y))

filtered
```

```python
# Create a new column for the start location
filtered[['x_start', 'y_start']] = pd.DataFrame(filtered.location.tolist(), index=filtered.index)

# Create a new column for the end location
filtered[['x_end', 'y_end']] = pd.DataFrame(filtered.pass_end_location.tolist(), index=filtered.index)

filtered
```

```python
# Set the pitch dimensions
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white')

# Create a figure
fig, ax = pitch.draw()

# Plot the start and end locations
pitch.scatter(filtered.x_start, filtered.y_start, s=100, c='blue', ax=ax)
pitch.scatter(filtered.x_end, filtered.y_end, s=100, c='red', ax=ax)

# Plot the pass lines
for i in range(len(filtered)):
    pitch.arrows(filtered.x_start[i], filtered.y_start[i], filtered.x_end[i], filtered.y_end[i], width=2, color='blue', ax=ax)

# Add a title
ax.set_title('Lionel Messi Passes vs NY Red Bulls 2023', fontsize=20)

# Show the plot
fig.set_size_inches(12, 8)
fig.tight_layout()
fig.show()
```

Create a map of every pass made during that game.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a dictionary to map player IDs to player names
player_names = {
    5503: 'Lionel Messi',
    222155.0: 'D. Amarilla',
    32878.0: 'F. Farias',
    5203.0: 'S. Busquests',
    36430.0: 'D. Espinosa',
    3658.0: 'D. Yedlin',
    5211.0: 'J. Alba',
    225209.0: 'B. Cremaschi',
    36149.0: 'R. Belmar'
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes for each player
for player_id in filtered['player_id'].dropna().unique():
    G.add_node(player_id)

# Add edges for each pass with weights
for i in range(len(filtered)):
    if pd.notna(filtered['player_id'][i]) and pd.notna(filtered['pass_recipient_id'][i]):
        if G.has_edge(filtered['player_id'][i], filtered['pass_recipient_id'][i]):
            G[filtered['player_id'][i]][filtered['pass_recipient_id'][i]]['weight'] += 1
        else:
            G.add_edge(filtered['player_id'][i], filtered['pass_recipient_id'][i], weight=1)

# Create a layout for the nodes
pos = nx.spring_layout(G)

# Draw the nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=700)
edges = nx.draw_networkx_edges(G, pos, width=[G[u][v]['weight'] for u, v in G.edges()])

# Use the player_names dictionary to label the nodes
nx.draw_networkx_labels(G, pos, labels=player_names, font_size=12, font_color='Orange')

# Show the plot
plt.axis('off')
plt.show()

```

Create Pass Network Graph

```python
import pandas as pd
from mplsoccer import Pitch, Sbopen

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

MATCH_ID = 3877060
```

```python
# Initialize the StatsBomb parser
parser = Sbopen()
```

```python
# Get the match data
pass_network_event_df, related_2, freeze_2, tactics_2 = parser.event(3877060)
```

```python
# Filter for Red Bulls events
pass_network_event_df = pass_network_event_df[pass_network_event_df['team_name'] == 'Inter Miami']
```

```python
# filter for pass actions
red_bull_passes = pass_network_event_df[pass_network_event_df['type_name']=='Pass']
filtered_rb_passes = red_bull_passes[['player_name','x','y','end_x', 'end_y','pass_recipient_name','outcome_id','outcome_name']]
```

```python
filtered_rb_passes[filtered_rb_passes['outcome_name'].isnull()]
```

```python
pass_network_event_df
```

```python
# Get the lineup data and extract jersey numbers
lineup_df = parser.lineup(MATCH_ID)
jersey_data=lineup_df[['player_id','jersey_number']]

jersey_data
```

```python
# Merge the jersey numbers with the evenmt data
final_df= pd.merge(pass_network_event_df,jersey_data,on='player_id')
final_df
```

```python
# make player_id the jersey number
final_df['passer'] = final_df['jersey_number']
passes = final_df[final_df['type_name']=='Pass']
successful = passes[passes['outcome_name'].isnull()]
successful
```

```python
# Finalized DataFrame
rec= pd.to_numeric(successful['pass_recipient_id'],downcast='integer')
jersey_data.rename(columns={'player_id':'pass_recipient_id'},inplace=True)
jersey_data.rename(columns={'jersey_number':'pass_recipient'},inplace=True)
successful=pd.merge(final_df,jersey_data,on='pass_recipient_id')
successful
```

```python
# Get the first substitution minute
subs = event_df[event_df['type_name']=='Substitution']
subs = subs['minute']
firstSub=subs.min()

# Filter for passes before first substitution
successful = successful[successful['minute']<firstSub]
```

```python
# Print variable types
successful.dtypes
```

```python
# Get the average location of each player
average_locations= successful.groupby('passer').agg({'x':['mean'],'y':['mean','count']})
average_locations.columns = ['x','y','count']
average_locations
```

```python
# Make pass_between dataframe
pass_between = successful.groupby(['passer','pass_recipient']).id.count().reset_index()
pass_between.rename({'id':'pass_count'},axis='columns',inplace=True)
pass_between
```

```python
# Merge dataframes pass_between and average_locations
pass_between= pass_between.merge(average_locations,left_on='passer',right_index=True)
pass_between
```

```python
# Merge dataframes pass_between and average_locations on pass_recipient
pass_between= pass_between.merge(average_locations,left_on='pass_recipient',right_index=True,suffixes=['','_end'])
pass_between
```

```python
# Filter out passes with no count, low count, or duplicates
# pass_between =pass_between[pass_between['pass_count']>3]
pass_between.drop_duplicates()
pass_between
```

Finally we can create the graph!

```python
# Create a pitch
pitch = Pitch(pitch_color='grass', line_color='#B6BBC4')

# Set the figure size
fig, ax = pitch.draw(figsize=(8, 6), constrained_layout=True, tight_layout=False)

# Set face color
fig.patch.set_facecolor('#FFFFFF')

# Draw lines and nodes
pass_lines = pitch.lines(1.2*pass_between.x, 0.8*pass_between.y, 1.2*pass_between.x_end, 0.8*pass_between.y_end, lw=pass_between.pass_count*0.5,color='pink', zorder=1, ax=ax)
nodes = pitch.scatter(1.2 * average_locations.x, 0.8 * average_locations.y, s=20*average_locations['count'].values, color='white', edgecolors='#a6aab3', linewidth=2, alpha=1, zorder=1, ax=ax)

# Annotate average_locations
for index, row in average_locations.iterrows():
    pitch.annotate(index, xy=(1.2 * row.x, 0.8 * row.y), c='#161A30',fontweight='bold', va='center', ha='center', size=8, ax=ax)

# Add the title
ax.set_title('Inter Maimi Pass Network vs NY Red Bulls (Aug-2023)', color='red', va='center', ha='center', fontsize=12,fontweight='bold',pad=20,loc='center')

plt.show()

```
